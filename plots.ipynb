```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(context='talk', font_scale=1.0, color_codes=True, palette='deep', style='ticks',
rc={'mathtext.fontset': 'cm', 'xtick.direction': 'in','ytick.direction': 'in',
            'axes.linewidth': 1.5, 'figure.dpi':70, 'text.usetex':False})

plt.rcParams['font.family'] = 'Arial'

%load_ext autoreload
%autoreload 2
```

```python
niter = 10000
Ne = 256

graph_type = "bernoulli"
Ns = [20,50,75]
ps = [0.2,0.1,0.1]
for N,p in zip(Ns,ps):
    prefix = "./outs/I%08d_N%d_p%.3f_Ne%d_G" %(niter,N,p,Ne) + graph_type

    nruns = 6

    losses_all = []
    bias1s_all = []
    bias2s_all = []

    plt.close("all")
    fig,axis = plt.subplots(3,1,figsize=(6,10))
    for run in range(nruns):
        # Load the npz file
        data = np.load(prefix + "/iter%d.npz"%run)
        losses = data['arr_0']
        bias1s = data['arr_2']
        bias2s = data['arr_3']
        best_loss = data['arr_4'][0]

        axis[0].plot(losses, 'C%d-'%run, alpha = 0.4)
        axis[0].plot([0,len(losses)],[best_loss,best_loss], 'C%d--'%run, alpha = 0.4)
        axis[1].plot(np.mean(bias1s,axis=1), 'k-', alpha = 0.3)
        axis[2].plot(np.mean(bias2s,axis=1), 'k-', alpha = 0.3)

        losses_all.append(losses)
        bias1s_all.append(np.mean(bias1s,axis=1))
        bias2s_all.append(np.mean(bias2s,axis=1))

    #axis[0].plot(np.mean(losses_all,axis=0), 'k-', lw=2)
    axis[1].plot(np.mean(bias1s_all,axis=0), 'k-', lw=2)
    axis[2].plot(np.mean(bias2s_all,axis=0), 'k-', lw=2)

    axis[0].set_ylabel('Loss')
    axis[0].set_xlabel('Iterations (x100)')

    axis[1].set_ylabel(r'$\log p_{opt}/p_{subopt}$')
    axis[1].set_xlabel('Iterations (x100)')
    axis[1].set_ylim(-0.1,0.8)

    axis[2].set_ylabel(r'$p_{opt} - p_{subopt}$')
    axis[2].set_xlabel('Iterations (x100)')
    axis[2].set_ylim(-0.1,0.25)

    fig.suptitle(graph_type + ' N=%d, p=%.2f'%(N,p))
    fig.tight_layout()

    fig.savefig("figs/losses_and_biases_%s_N%d_p%.2f_Ne%d.pdf"%(graph_type,N,p,Ne))
    plt.show()
    
```

```python
niter = 15000
Nes = [8,16, 24, 32, 48, 64, 80, 96, 112, 128]
N = 31
p = 0.1
graph_type = "spoke"
plt.close("all")
fig,axis = plt.subplots(3,1,figsize=(5,8))
for Ne in Nes:
    prefix = "./outs/I%08d_N%d_p%.3f_Ne%d_G" %(niter,N,p,Ne) + graph_type

    nruns = 6

    losses_all = []
    bias1s_all = []
    bias2s_all = []
    for run in range(nruns):
        # Load the npz file
        data = np.load(prefix + "/iter%d.npz"%run)
        losses = data['arr_0']
        bias1s = data['arr_2']
        bias2s = data['arr_3']
        best_loss = data['arr_4'][0]

        losses_all.append(losses[-1])
        bias1s_all.append(np.mean(bias1s,axis=1)[-1])
        bias2s_all.append(np.mean(bias2s,axis=1)[-1])

    axis[0].errorbar([Ne],np.mean(losses_all), yerr= np.std(losses_all)/np.sqrt(nruns),  fmt = 'o', color = 'k')
    axis[0].plot([Nes[0],Nes[-1]],[best_loss,best_loss], 'k--', alpha = 0.4)

    axis[1].errorbar([Ne],np.mean(bias1s_all), yerr= np.std(bias1s_all)/np.sqrt(nruns),  fmt = 'o', color = 'k')

    axis[2].errorbar([Ne],np.mean(bias2s_all), yerr= np.std(bias2s_all)/np.sqrt(nruns),  fmt = 'o', color = 'k')


#fig.suptitle(graph_type + ' Ne=%d'%(Ne))
axis[0].set_ylabel('Loss')
axis[0].set_xlabel(r"$N_e$")
axis[0].set_ylim(0, 0.2)

axis[1].set_ylabel('Bias 1')
axis[1].set_xlabel(r"$N_e$")

axis[2].set_ylabel('Bias 2')
axis[2].set_xlabel(r"$N_e$")

fig.tight_layout()
fig.savefig("figs/bias_Ne_dependence_%s_N%d_p%.2f.pdf"%(graph_type,N,p))
plt.show() 
    
```
